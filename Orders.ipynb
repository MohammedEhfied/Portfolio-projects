```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
```

```python
# reading the data

df = pd.read_csv(r'C:\Users\lcs\Downloads\New folder\NewOrders.csv')
df.head()
```

# Data cleaning

```python
# show all the columns

pd.set_option('display.max_columns', None)
df.head()
```

```python
df.info()
```

```python
# make 'Row ID' and 'Discount' columns better

df['Row ID'] = np.arange(0,len(df))

df['Discount'] = df['Discount'] * 100
df['Discount'] = df['Discount'].astype(int) 
df.rename(columns = {'Discount':'Discount (%)'}, inplace = True)

df.head()
```

```python
# adding 'Year Date' column

year = list()

for i in df['Order Date']:
    str(i)
    year.append(i[-4:])

df.insert(3,'Order Year',year)

df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {} %'.format(col,pct_missing))
```

```python
# lets see how many rows are missing in 'Product Base Margin' column

df[['Product Base Margin','Row ID']].isnull().groupby(by='Product Base Margin').count()
```

```python
# filling missing values with zeros

df['Product Base Margin'] = df['Product Base Margin'].fillna(0)
```

```python
# dropping duplicates if exist

df.drop_duplicates()
```

# Data Exploration

```python
# Correlation Matrix between all numeric columns

df.corr(method ='pearson')
```

```python
df.corr(method ='spearman')
```

```python
df.corr(method ='kendall')
```

```python
# change object values to numeric

df_numerized = df.copy(deep=True)


for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
sns.heatmap(df_numerized.corr(), linewidths=1, annot=True)

plt.title("Correlation matrix for all columns \n")

plt.show()
```

```python
# Looking at Product Categories by profit revenue

Category = df.groupby('Product Category')[["Profit"]].sum()

Category = Category.sort_values('Profit', ascending = False)

Category = Category['Profit'].astype('int64') 

Category
```

```python
# top 5 sold products

Product_Name = df.groupby('Product Name')[["Order Quantity"]].sum()

Product_Name = Product_Name.sort_values('Order Quantity', ascending = False)[:5]

Product_Name = Product_Name['Order Quantity'].astype('int64') 

Product_Name
```
